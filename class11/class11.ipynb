```python
# In python this pointer =  self

class StudentCard():
    def __init__(self,age:int,rollno:int,name:str) -> None:
        if(age < 18 or age > 60):
            raise StudentAgeError("The age < 18 or age > 60 is not allowed")
        self.age = age
        self.rollno = rollno
        self.name = name

            
class StudentAgeError(Exception):
    pass
        
```

```python
# By default file mode is "r" -> read
# By default "rt" mode read and text mode open, rb means read mode and byte mode 
# w -> truncate and write on the empty file. w+ -> truncate, write on the empty file and read from start
# a -> append,write at end , donot truncate , a+ -> append, write at end, donot truncate,and read 
# r+ -> read, and write at start, do not truncate 
# truncation only happens with w or w+ or w modes
# Second parameter in open represents mode

with open("./abc.txt","r") as file:
    print(file.readlines()[:5])
```

```python
with open("./abc.txt","r") as file:
    print(file.read())
    file.write("haseeb") # will throw exception since we didnt give write access
```

```python
with open("./abc1.txt","w") as file:
    file.write("Haseeb IJaz")
    print(file.read()) # raises exception since we didnt have read permissions
    
```

```python
# With every read and write our cursor moves
#After reading , it will write where it ended reading

with open('./abc1.txt',"r+") as file:
    print(file.read()) # takes the cursor to the end of file
    file.write("\n next line") 
```

```python
with open("./abc1.txt","r+") as file:
    print(file.read()) #cursor moves to the end of the file after reading the whole file
    file.write("Haseeb") #cursor writes at the end of file where reading was left with cursor
    print(file.read()) #cursor reads from where writing was left which is EOF so nothing is read
```

```python
with open("./abc1.txt","r+") as file:
    print(file.read()) #cursor moves to the end of the file after reading the whole file
    file.write("Haseeb") #cursor writes at the end of file where reading was left with cursor
    file.seek(0) #cursor moves to position 0 index
    print(file.read()) #cursor reads from 0 index and so reads the whole file
```

```python
from typing import List
import PyPDF2

def read_pdf(path:str) -> List[str]:
    with open(path,"rb") as file:
        reader: PyPDF2.PdfReader = PyPDF2.PdfReader(file)
        text_content:List[str] = [page.extract_text() for page in reader.pages]
        return text_content
read_pdf("sample.pdf")
```
